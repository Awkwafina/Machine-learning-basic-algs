```python
import numpy as np
from sklearn.datasets.samples_generator import make_blobs
from matplotlib import pyplot as plt
from pandas import DataFrame
```

```python
class Perceptron:
    def __init__(self):
        self.weights = np.zeros(len(X[0])+1)
        self.j = 0
    def fit(self, X, y):
        is_missclassified = True
        while is_missclassified:
            is_missclassified = False
            for i, x in enumerate(X):
                if self.predict(x)!=y[i]:
                    is_missclassified = True
                    x = np.insert(x, 0, 1)
                    self.weights = self.weights + y[i]*x
                    self.j+=1
        print(f'Number of iterations taken for convergence is {self.j}')
        print(f'the set of weights is  {self.weights}')
        return self.weights
        
    def predict(self, x):
        return np.sign(np.dot(np.transpose(self.weights[1:]),x) + self.weights[0])
    
    def check_prediction(self, X, y):
        for i,x in enumerate(X):
            if self.predict(x) != y[i]:
                break
            else:
                print('SUCCESS!')
        
```

```python
#I am implementing this function to follow the book (it is either -1 or +1)
def change_zero(y):
    for i in range(0, len(y)):
        if y[i]==0:
            y[i]=-1
    return y
    
```

## part A

```python
# generate 2d classification dataset
X, y = make_blobs(n_samples=20, centers=2, n_features=2)
print(X)
y = change_zero(y)
print(y)
c0 = y==-1
c1 = y==1
print(c0)
plt.xlabel('x1', fontsize = 16)
plt.ylabel('x2', fontsize = 16)
plt.plot(X[:,0][c0], X[:,1][c0], 'o', mec='r', mfc='none')
plt.plot(X[:,0][c1], X[:,1][c1], 'o', mec='g', mfc='none')
```

## Part B

```python
p = Perceptron()
w = p.fit(X, y)
x2 = [(-w[0]-w[1]*x[0])/w[2] for x in X ]
plt.plot(X[:,0],x2, label = 'hypothesis g')
target_f = [6+((0-6)/(7-0))*x[0] for x in X]
plt.plot(X[:, 0], target_f,label = 'target f')
plt.xlabel('x1', fontsize = 16)
plt.ylabel('x2', fontsize = 16)
plt.legend(loc='upper left')
plt.plot(X[:,0][c0], X[:,1][c0], 'o', mec='r', mfc='none')
plt.plot(X[:,0][c1], X[:,1][c1], 'o', mec='g', mfc='none')
plt.show()
```

#### Comment:
Since the number of iterations before convergence is 2 here, two lines are very far from each other.
'Closeness' really depends on number of iterations.
